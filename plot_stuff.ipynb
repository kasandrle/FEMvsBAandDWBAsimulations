```python
import numpy as np

from os import listdir
from os.path import isfile, join, expanduser
import pandas as pd
import pickle
import sys

import matplotlib.pyplot as plt
```

```python
path = '.'
onlyfiles = [f for f in listdir(path) if isfile(join(path, f)) and f.endswith('.csv') and f.startswith('e')]

exp_data_all = []
for file in onlyfiles[:]:
    #print(file)
    file = join(path, file)
    loaded_dict = pd.read_csv(file)  
    
    exp_data_all.append(loaded_dict)
```

```python
exp_data_all[0].dtypes

for i,entry in enumerate(exp_data_all):
    #print(entry.dtypes)
    plt.plot(entry['Kx [1/m]'],entry['intensity'],label=onlyfiles[i])
plt.legend()   
plt.yscale('log')
```
